# application of diffusion map to the swiss roll

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_swiss_roll
```

```python
from diffusion_maps_public import*
```

## create example data set

```python
#creating swissroll
X, color = make_swiss_roll(n_samples=2000, noise=0.01, hole=False)
```

```python
#show created data
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=color)
plt.show()
```

## apply diffusion map

```python
dmap = diffusion_map(X,epsilon = 0.1, N = 'all', t = 1, standardize_data = True, num_eigenv = 3)
```

## plot results

```python
#show created data
plt.figure()
plt.scatter(dmap.dmap[:, 0], dmap.dmap[:, 1], c=color)
plt.xlabel("diffusion component 1")
plt.ylabel("diffusion component 2")
plt.show()
```
